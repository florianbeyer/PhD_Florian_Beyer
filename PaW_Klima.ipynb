```python
wlevel_10_17 = wlevel.copy()
wlevel_10_17 = wlevel_10_17.drop(columns=['2018'])

wlevel_18 = pd.DataFrame()
wlevel_18['2018'] = wlevel['2018']

wlevel_plot = pd.DataFrame()
wlevel_plot['2018'] = wlevel_18['2018']
wlevel_plot['min'] = wlevel_10_17.min(axis=1)
wlevel_plot['max'] = wlevel_10_17.max(axis=1)


# Set the locator
locator = mdates.MonthLocator()  # every month
# Specify the format - %b gives us Jan, Feb...
fmt = mdates.DateFormatter('%b')

plt.rcParams.update({'font.size': 22})

fig = plt.figure(figsize=(12,12))
ax = plt.axes()

numdays = 365
base = datetime.datetime(2018, 1, 1, 0, 0, 0, 0)
x = [base + datetime.timedelta(days=x) for x in range(0,numdays)]

ax.fill_between(x, wlevel_plot['min'], wlevel_plot['max'], color='lightgray', label='2010-2017')
ax.plot(x, wlevel_10_17, color='grey', linewidth=0.3)
ax.plot(x, wlevel_plot['2018'], color='black', label='2018');
plt.axhline(y=0, xmin=0, xmax=1, linewidth=1, color = 'k', linestyle='dashed')


#point = pd.DataFrame({'x': [datetime.datetime(2018, 4, 8,0, 0, 0, 0)], 'y':[0.65]})
# point.plot(x='x', y='y', ax=ax)
#plt.plot(point['x'], point['y'], marker='v', markersize=15, color="red")
ax.legend(loc='lower left')

ax.set_ylabel('Wasserstand [m] Ã¼ber GelÃ¤ndeoberkante')
#ax.set_xlabel('DOY')

X = plt.gca().xaxis
X.set_major_locator(locator)
# Specify formatter
X.set_major_formatter(fmt)


plt.savefig('R:/OwnCloud/DissBeyer/Diss2020/Arbeit/Abbildungen/05_Ergebnisse/paw_wasserstand.png', dpi=300, quality=95, format="png")

plt.show()
```
